```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Path where the model was saved
model_save_path = "/kaggle/input/gpt2-pt4/results/checkpoint-25000"

# Load the trained tokenizer and model with the same settings as during training
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-medium', bos_token='', eos_token='', pad_token='')

# Load the model
model = GPT2LMHeadModel.from_pretrained(model_save_path)

# Ensure your model is in evaluation mode
model.eval()
```

```python
# Generate samples
prompt = "I think these taste great! Especially when added to something else like a body armor drink. I like to drink these when I’m flying since the airplanes can be so drying. Is it helping? Honestly I don’t know because I don’t have the tools or education to properly measure that. However, feel like I don’t get as many dehydration headaches when I use these. So whether that’s the product working or just me drinking more fluids as a result of me using the product, I am more hydrated. <|summary|>"  # You can modify this prompt to suit your needs
encoding = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True, max_length=512)
input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]
if torch.cuda.is_available():
    input_ids = input_ids.to("cuda")
    attention_mask = attention_mask.to("cuda")

sample_outputs = model.generate(
    input_ids,
    attention_mask=attention_mask,
    do_sample=True,
    top_k=50,
    max_new_tokens=20,
    top_p=0.95,
    temperature=1.9,
    num_return_sequences=5
)

# Display the generated samples
for i, sample_output in enumerate(sample_outputs):
    print(f"{i+1}: {tokenizer.decode(sample_output, skip_special_tokens=True)}\n")
```
